```python
from qiskit import * #importing essential libraries
import numpy as np
%matplotlib inline
```

```python
# Building a simple Quantum circuit of 3 qubits or registers
circ = QuantumCircuit(3)

#Add Hadamard gate for superposition
circ.h(0)

#add controlled not gate for target qubit 1 from control qubit 0
circ.cx(0,1)

#add controlled not gate for target qubit 2 from control qubit 0
circ.cx(0,2)
```

```python
circ.draw('mpl')
```

```python
from qiskit import Aer
backend = Aer.get_backend('statevector_simulator')
```

```python
job = execute(circ, backend = backend)
```

```python
result = job.result()
```

```python
output = result.get_statevector(circ, decimals = 3)
print(output)
```

```python
from qiskit.visualization import plot_state_city
plot_state_city(output)
```
